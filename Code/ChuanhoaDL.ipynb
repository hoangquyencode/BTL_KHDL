```python
import pandas as pd
import numpy as np

# Thiết lập seed để dữ liệu ngẫu nhiên nhưng cố định quy luật
np.random.seed(42)

nam_list, thang_list, vung_list, loai_che_list = [], [], [], []
san_luong_list, gia_ban_list, luong_mua_list, nhiet_do_list, dien_tich_list = [], [], [], [], []

vung_trong = ['Tan Cuong', 'Dai Tu', 'Dong Hy', 'Phu Luong']
loai_che = ['Che Dinh', 'Che Non Tom', 'Che Moc Cau', 'Che Bup']

# Vòng lặp quét qua từng Năm -> từng Tháng -> từng Vùng -> từng Loại chè (960 dòng)
for nam in range(2021, 2026):
    for thang in range(1, 13):
        # Logic Thời tiết chung
        if thang in [5, 6, 7, 8]:  # Mùa hè
            luong_mua_nen = np.random.randint(220, 320)
            nhiet_do_nen = round(np.random.uniform(29.0, 33.0), 1)
            he_so_san_luong = 1.8
        elif thang in [11, 12, 1, 2]:  # Mùa đông
            luong_mua_nen = np.random.randint(20, 45)
            nhiet_do_nen = round(np.random.uniform(15.5, 19.0), 1)
            he_so_san_luong = 0.6
        else:  # Mùa xuân, thu
            luong_mua_nen = np.random.randint(90, 150)
            nhiet_do_nen = round(np.random.uniform(22.0, 26.0), 1)
            he_so_san_luong = 1.2

        for vung in vung_trong:
            dt_dict = {'Tan Cuong': 450, 'Dai Tu': 1650, 'Dong Hy': 930, 'Phu Luong': 1150}
            dien_tich = dt_dict[vung] + np.random.randint(-10, 15)

            luong_mua = luong_mua_nen + np.random.randint(-10, 10)
            nhiet_do = round(nhiet_do_nen + np.random.uniform(-0.5, 0.5), 1)

            for loai in loai_che:
                base_san_luong = {'Che Dinh': 0.01, 'Che Non Tom': 0.03, 'Che Moc Cau': 0.06, 'Che Bup': 0.09}
                san_luong = (dien_tich * base_san_luong[loai] * he_so_san_luong) * np.random.uniform(0.9, 1.1)
                san_luong = round(san_luong, 2)

                he_so_thuong_hieu = 1.4 if vung == 'Tan Cuong' else 1.0
                base_gia = {'Che Dinh': 3000000, 'Che Non Tom': 1200000, 'Che Moc Cau': 350000, 'Che Bup': 180000}

                gia = base_gia[loai] * he_so_thuong_hieu * (1 + (nam - 2021) * 0.06)
                if thang in [12, 1]:
                    gia *= 1.20
                gia = int(gia * np.random.uniform(0.95, 1.05))

                nam_list.append(nam)
                thang_list.append(thang)
                vung_list.append(vung)
                loai_che_list.append(loai)
                san_luong_list.append(san_luong)
                gia_ban_list.append(gia)
                luong_mua_list.append(luong_mua)
                nhiet_do_list.append(nhiet_do)
                dien_tich_list.append(dien_tich)

df_master = pd.DataFrame({
    'Nam': nam_list, 'Thang': thang_list, 'Vung_Trong': vung_list, 'Loai_Che': loai_che_list,
    'San_Luong_Tan': san_luong_list, 'Gia_Ban_VND_Kg': gia_ban_list, 'Luong_Mua_mm': luong_mua_list,
    'Nhiet_Do_C': nhiet_do_list, 'Dien_Tich_Ha': dien_tich_list
})

df_master['Doanh_Thu_Trieu_VND'] = round((df_master['San_Luong_Tan'] * 1000 * df_master['Gia_Ban_VND_Kg']) / 1000000, 2)

# Lệnh quan trọng nhất: Xuất dữ liệu ra file lưu vào Colab
df_master.to_csv('du_lieu_960_dong_toan_dien.csv', index=False, encoding='utf-8-sig')
print("Đã tạo file thành công trong bộ nhớ Colab!")
```
